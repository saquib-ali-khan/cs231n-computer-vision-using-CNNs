###########  np.argsort ###############

One dimensional array:
>>>

>>> x = np.array([3, 1, 2])
>>> np.argsort(x)
array([1, 2, 0])

Two-dimensional array:
>>>

>>> x = np.array([[0, 3], [2, 2]])
>>> x
array([[0, 3],
       [2, 2]])

>>>

>>> np.argsort(x, axis=0)
array([[0, 1],
       [1, 0]])

>>>

>>> np.argsort(x, axis=1)
array([[0, 1],
       [0, 1]])



```python
############## np.split ####################
#returns a list containing the elements of which point to arrays that have been formed due to the split
#accepts paramters, as the array to be splitted , and then a no. maybe , or a list and then the axis of split (default=0)
# as in examples below
import numpy as np
a=np.arange(10)

b=np.split(a,2)

print b
```

```python
c=np.split(a,[2,4,8])  # second arguent is a list of partition points
print c
```

```python
### now lets say the array given to split is a 2-D array
a=np.array(([2,3,4],[1,2,3],[5,6,7],[5,8,9]))
b=np.split(a,2)
print b
```

## the output was 2 arrays split across the rows , because default axis of np.split is axis=0

```python
a=np.array(([1,2,3,4,5,4],[2,3,5,6,7,8],[5,6,8,3,4,5]))
b=np.split(a,3,axis=1)
print b
#########output is 3 , 3x2 arrays due to vertical split of array a.
```

```python
b=np.split(a,[3,5],axis=1)
print b
####### output is 3 arrays of sizes 3x3 , 3x2, 3x1  , i.e after vertical split of a at points 3 and 5 
```

```python
##################### np.concatenate #################################
# concatenates given a sequence of arrays along a given axis 
# and returns a concatenated array
# as follows

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6]])
c = np.concatenate((a, b), axis=0)
d = np.concatenate((a, b.T), axis=1)
print  "c is : \n",c
print  "d is : \n",d 
```

```python
print np.concatenate([a,b,c])
```
